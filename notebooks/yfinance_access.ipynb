### Getting to know basics in `yfinance`
 `yfinance` is a library which can be used to retrieve market-data from Yahoo Finance. 

```python
import yfinance as yf
```

```python
# Apple Stock as example
apple_ticker = "AAPL"
apple_stock = yf.Ticker(apple_ticker)
```

```python
apple_stock.info
```

```python
apple_stock.history(period="max")
```

```python
apple_stock.splits
```

```python
apple_stock.institutional_holders
```

```python
apple_stock.recommendations
```

```python
apple_stock.financials
```

```python
apple_stock.actions
```
