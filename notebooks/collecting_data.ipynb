# P/E ratio

```python
# pick stock
stock_tickers = ["AAPL", "NVDA", "GOOGL"]

# find todays date with datetime
today = datetime.now().date()

# format and delete leading zeroes
today_day = today.strftime("%d")
today_month = today.strftime("%m")
today_year = today.strftime("%Y")

today_day = int(str(today_day))
today_month = int(str(today_month))
today_year = int(str(today_year))


# set start and end date of timespan
end_date = datetime(year=today_year, month=today_month, day=today_day)
start_date = datetime(year=2026, month=1, day=1)


df_multi = yf.download(
    tickers=stock_tickers,
    start=start_date,
    end=end_date,
    interval="1d",
    group_by="ticker"
)
df_multi

# access just AAPL
#df_multi["AAPL"]

# access just AAPL closing
#df_multi["AAPL"]["Close"]
```

```python
# initialize a new figure
plt.figure(figsize=(14, 7))
sns.set(style="whitegrid")

# loop over the tickers
for ticker in stock_tickers:
    # plot the closing price for each
    sns.lineplot(
        data=df_multi[ticker]["Close"], #IMPORTANT LINE
        label=ticker,
        linewidth=2
    )

# set the plot title
plt.title(
    f"Stock Closing Prices ("
    f"{start_date.strftime('%Y-%m-%d')} "
    f"to {end_date.strftime('%Y-%m-%d')})"
)

# set the plot labels
plt.xlabel("Date")
plt.ylabel("Closing Price ($)")

# finish constructing the plot
plt.tight_layout()
plt.show()
```
